```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# List of duplicate image pairs (as tuples)
image_pairs = [
    ("/content//WildRF/val/1_fake/0017.jpg",
     "/content/WildRF/train/1_fake/0236.jpg"),

    ("/content/WildRF/val/1_fake/0055.png",
     "/content/WildRF/train/1_fake/0476.png"),

    ("/content//WildRF/test/reddit/1_fake/0140.png",
     "/content//WildRF/train/1_fake/0383.png"),

    ("/content/WildRF/test/reddit/1_fake/0242.jpg",
     "/content/WildRF/train/1_fake/0533.jpg"),

    ("/content/WildRF/test/reddit/1_fake/0243.jpg",
     "/content/WildRF/val/1_fake/0065.jpg"),
]



num_pairs = len(image_pairs)

# 4 rows: A image, B image, A path, B path; num_pairs columns
fig, axes = plt.subplots(4, num_pairs, figsize=(4 * num_pairs, 10))

for col, (img_a_path, img_b_path) in enumerate(image_pairs):
    # Load images
    img_a = mpimg.imread(img_a_path)
    img_b = mpimg.imread(img_b_path)

    # Row 0: Image A
    axes[0, col].imshow(img_a)
    axes[0, col].axis('off')

    # Row 2: Path A
    axes[1, col].axis('off')
    axes[1, col].text(0.5, 0.5, img_a_path, ha='center', va='center', wrap=True, fontsize=12)

    # Row 1: Image B
    axes[2, col].imshow(img_b)
    axes[2, col].axis('off')

    # Row 3: Path B
    axes[3, col].axis('off')
    axes[3, col].text(0.5, 0.5, img_b_path, ha='center', va='center', wrap=True, fontsize=12)

plt.tight_layout(h_pad=2)
plt.show()
```

#Visualising random photos from training fake and training real

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

# Define paths
fake_path = "/content/WildRF/train/1_fake"
real_path = "/content/WildRF/train/0_real"

# Sample 5 random image filenames
fake_images = random.sample(os.listdir(fake_path), 5)
real_images = random.sample(os.listdir(real_path), 5)

# Set up figure and axes
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

# Plot real images (top row)
for i, img_name in enumerate(real_images):
    img_path = os.path.join(real_path, img_name)
    img = Image.open(img_path)
    axes[0, i].imshow(img)
    axes[0, i].axis('off')
    if i == 0:
        axes[0, i].set_ylabel("Real", rotation=0, fontsize=14, labelpad=60, va='center')

# Plot fake images (bottom row)
for i, img_name in enumerate(fake_images):
    img_path = os.path.join(fake_path, img_name)
    img = Image.open(img_path)
    axes[1, i].imshow(img)
    axes[1, i].axis('off')
    if i == 0:
        axes[1, i].set_ylabel("Fake", rotation=0, fontsize=14, labelpad=60, va='center')

# Adjust spacing
plt.subplots_adjust(left=0.08, wspace=0.05, hspace=0.05)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from PIL import Image
import textwrap

# Set how many pairs you want per A4 page
pairs_per_page = 6

# A4 size in inches
a4_width, a4_height = 8.27, 11.69

# Function to plot image pairs
def plot_pairs(duplicate_pairs, output_pdf_path):
    pages = (len(duplicate_pairs) + pairs_per_page - 1) // pairs_per_page

    with PdfPages(output_pdf_path) as pdf:
        for page in range(pages):
            fig, axs = plt.subplots(pairs_per_page, 2, figsize=(a4_width, a4_height))
            fig.subplots_adjust(wspace=0.05, hspace=0.8)  # extra spacing for wrapped text

            start_idx = page * pairs_per_page
            end_idx = min(start_idx + pairs_per_page, len(duplicate_pairs))
            pairs_on_this_page = end_idx - start_idx

            for idx in range(pairs_per_page):
                if idx >= pairs_on_this_page:
                    axs[idx, 0].axis('off')
                    axs[idx, 1].axis('off')
                    continue

                pair = duplicate_pairs[start_idx + idx]
                for j in range(2):
                    img_path = pair[j]
                    ax = axs[idx, j]
                    ax.axis('off')
                    try:
                        img = Image.open(img_path)
                        ax.imshow(img)

                        # Wrap the path string into multiple lines
                        wrapped_path = "\n".join(textwrap.wrap(img_path, width=60))
                        ax.set_title(wrapped_path, fontsize=5, pad=10, loc='center')

                    except Exception as e:
                        ax.text(0.5, 0.5, 'Error\nLoading\nImage', ha='center', va='center', fontsize=6)
                        print(f"Error loading {img_path}: {e}")

            pdf.savefig(fig)
            plt.close()


# Example usage
output_pdf_path = '/content/drive/MyDrive/Final_folder_code_thesis/ægteduplicates_visualized.pdf'
plot_pairs(duplicate_pairs, output_pdf_path)

```
